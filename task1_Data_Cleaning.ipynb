### Diffrent methods to handle missing values in data

```python
#Dropping null values

m1=df.dropna(inplace=True)
```

```python
df.dropna(how="all") #If the row contains all the Null values then it drops it 
```

```python
df.dropna(how="any") #if row contains any Null values then it drops it
```

```python
df.fillna(0) #It replaces all the Null values with zeroes
```

```python
df.fillna(method='ffill') #FORWARD FILLING
```

```python
df.fillna(method='bfill')#Bacward Filling
```

```python
df.fillna(method="pad") #Pad=previous value (same as Forward feeling)
```

```python
df.fillna(method="pad",axis=1)  #Filling up the null values with previous value Column wise
```

```python
df.interpolate()  #Interpolate Method
```

```python
#Filling up the NUll values with Max
df.fillna(value=df['Cabin'].max())
```

```python
#Filling up the Null values with Minimum values

df.fillna(value=df['Age'].min())
```
